```python
import matplotlib.pyplot as plt
import seaborn as sns
paises = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
nombres = ['Norte America', 'USA', 'Japón', 'Otros']  # para títulos


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
axes = axes.flatten()  # aplanar la matriz de subplots

for i, pais in enumerate(paises):
    top10 = df[['Name', pais]].sort_values(pais, ascending=False).head(10)
    top10 = top10.sort_values(pais)  # para que el más vendido esté arriba

    top10.plot(kind='barh', x='Name', y=pais, ax=axes[i], color='steelblue', legend=False)
    axes[i].set_title(f'Top 10 Videojuegos - {nombres[i]}')
    axes[i].set_xlabel('Ventas')
    axes[i].set_ylabel('Videojuego')

plt.tight_layout()
plt.show()
```

```python
ultimos_años = sorted(df['Year'].unique())[-5:]  # los últimos 4 años disponibles
df_ultimos = df[df['Year'].isin(ultimos_años)]
df_ultimos
```

```python
ventas_genero = df_ultimos.groupby(['Year', 'Genre'], as_index=False)['Global_Sales'].sum()
pivot = ventas_genero.pivot(index='Genre', columns='Year', values='Global_Sales').fillna(0)

primer_año = pivot.columns[0]
ultimo_año = pivot.columns[-1]

pivot['crecimiento%'] = ((pivot[ultimo_año] - pivot[primer_año]) / pivot[primer_año]) * 100
mayor = pivot['crecimiento%'].idxmax()
menor = pivot['crecimiento%'].idxmin()

print(f"Género con mayor crecimiento: {mayor} (+{pivot.loc[mayor, 'crecimiento%']:.2f}%)")
print(f"Género con menor crecimiento: {menor} ({pivot.loc[menor, 'crecimiento%']:.2f}%)")

```

```python
ventas_plataforma = df_ultimos.groupby(['Year', 'Platform'], as_index=False)['Global_Sales'].sum()
pivot = ventas_plataforma.pivot(index='Platform', columns='Year', values='Global_Sales').fillna(0)

primer_año1 = pivot.columns[0]
ultimo_año1 = pivot.columns[-1]

pivot['crecimientop%'] = ((pivot[ultimo_año1] - pivot[primer_año1]) / pivot[primer_año1]) * 100
mayor1 = pivot['crecimientop%'].idxmax()
menor1 = pivot['crecimientop%'].idxmin()

print(f"Plataforma con mayor crecimiento: {mayor1} (+{pivot.loc[mayor1, 'crecimientop%']:.2f}%)")
print(f"Plataforma con menor crecimiento: {menor1} ({pivot.loc[menor1, 'crecimientop%']:.2f}%)")
```

```python
ventas_videoj = df_ultimos.groupby(['Year', 'Name'], as_index=False)['Global_Sales'].sum()
pivot = ventas_videoj.pivot(index='Name', columns='Year', values='Global_Sales').fillna(0)

primer_año2 = pivot.columns[0]
ultimo_año2 = pivot.columns[-1]

pivot['crecimientov%'] = ((pivot[ultimo_año2] - pivot[primer_año2]) / pivot[primer_año2]) * 100
mayor2 = pivot['crecimientov%'].idxmax()
menor2 = pivot['crecimientov%'].idxmin()

print(f"Videojuego con mayor crecimiento: {mayor2} (+{pivot.loc[mayor2, 'crecimientov%']:.2f}%)")
print(f"Videojuego con menor crecimiento: {menor2} ({pivot.loc[menor2, 'crecimientov%']:.2f}%)")
```
