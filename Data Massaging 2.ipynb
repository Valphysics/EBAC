```python
import pytz
from countryinfo import CountryInfo
def get_country_data(country_name):
    try:
        info=CountryInfo(country_name)
        capital = info.capital()
        timezones= info.timezones()
        timezone=timezones[0] if timezones else None
        return pd.Series([capital, timezone])
    except:
        return pd.Series([None, None])
df['Capital','Timezone']=df['Nationality'].apply(lambda x:get_country_data(x))
df
```
