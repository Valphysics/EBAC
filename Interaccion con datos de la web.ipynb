```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint  

headers = {
    'Access-Control-Allowed-Origin': '*',
    'Access-Control-Allowed-Methods': 'GET',
    'Access-Control-Allowed-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

url = 'https://www.mercadolibre.com.ar/'

try:
    data = requests.get(url, headers=headers, timeout=10)
    data.raise_for_status()  # Verificar si la solicitud fue exitosa
except requests.exceptions.RequestException as e:
    print(f"Error al obtener la pgina: {e}")
    data = None

url_data = []

if data:
    with open("mercadolibre.html", "w", encoding="utf-8") as f:
        f.write(data.text)

    html = BeautifulSoup(data.text, 'html.parser')

    articles = html.select('a[href*="/menu/categorias"], a[href*="/c/"]')

    for article in articles:
        category = article.get_text(strip=True)
        if category and category not in ['All', '', 'Categoras', 'Ms categoras']:  # Filtrar genricos
            url_data.append({'category': category})

    url_data = sorted(list({item['category']: item for item in url_data}.values()), key=lambda x: x['category'])

pprint(url_data)
```
